# ***Experiment-3***

**Clustering Task**

**1.Importing Libraries and Modules**

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import mode
```

**2.Loading Dataset**

```python
iris = load_iris()
X = iris.data
y = iris.target  # True labels (used only for evaluation)
```

**3.Applying K-means CLustering**

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

y_pred = kmeans.labels_
```

**4.Map Cluster Labels**

```python
label_map = {}

for i in range(3):
    mask = (y_pred == i)
    label_map[i] = mode(y[mask], keepdims=True)[0][0]

y_pred_mapped = np.array([label_map[label] for label in y_pred])
```

**5.Accuracy and Confusion Matrix**

```python
print("Clustering Accuracy:", accuracy_score(y, y_pred_mapped) * 100, "%")
cm = confusion_matrix(y, y_pred_mapped)
print("Confusion Matrix:\n", cm)
```

**6.Plot Confusion Matrix**

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering on Iris Dataset')
plt.show()
```
